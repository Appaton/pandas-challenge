```python
# Libraries
import pandas as pd
```

# Distric Summary

```python
# File to Load (Remember to Change These)
school_data_to_load = "csv/schools_complete.csv"
student_data_to_load = "csv/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
data.head(10)
```

```python
#Total Schools Calc
tot_schools = data['school_name'].nunique()
tot_schools = '{:,.0f}'.format(tot_schools)


# Total Students Calc
tot_students = data['Student ID'].nunique()
tot_students_format = '{:,.0f}'.format(tot_students)


# Total Budget Calc
unq_school = data.drop_duplicates('school_name',keep="first")
tot_budget = unq_school['budget'].sum()
tot_budget = '${:,.2f}'.format(tot_budget)

# Average Math Score Calc
avg_math = data['math_score'].mean()
avg_math = '{:,.2f}%'.format(avg_math)

# Average Reading Score Calc
avg_read = data['reading_score'].mean()
avg_read = '{:,.2f}%'.format(avg_read)


#Percent Passing Math Calc
unq_students = data.drop_duplicates('Student ID',keep='first')
index_fail1 = unq_students[unq_students['math_score']<70].index
unq_students.drop(index_fail1,inplace=True)
math_pass = unq_students['Student ID'].count()
per_math_pass ='{:,.2f}%'.format((math_pass/tot_students)*100)


# Percent Passing Reading Calc

unq_students = data.drop_duplicates('Student ID',keep='first')
index_fail2 = unq_students[unq_students['reading_score']<70].index
unq_students.drop(index_fail2,inplace=True)
read_pass = unq_students['Student ID'].count()
per_read_pass = '{:,.2f}%'.format((read_pass/tot_students)*100)


# Percent Overall Passing Calc

unq_students = data.drop_duplicates('Student ID',keep='first')
index_fail3 = unq_students[unq_students['reading_score']<70].index
unq_students.drop(index_fail3,inplace=True)
index_fail4 = unq_students[unq_students['math_score']<70].index
unq_students.drop(index_fail4,inplace=True)
rw_pass = unq_students['Student ID'].count()
per_rw_pass = '{:,.2f}%'.format((rw_pass/tot_students)*100)
```

```python
# Create DataFrame 

q1_df_dict = [{'Total Schools':tot_schools,
          'Total Students':tot_students_format,
          'Total Budget':tot_budget,
          'Average Math Score':avg_math,
          'Average Reading Score':avg_read,
          '% Passing Math':per_math_pass,
          '% Passing Reading':per_read_pass,
          '% Overall Passing':per_rw_pass}]
q1_df= pd.DataFrame(q1_df_dict)
q1_df
```

# School Summary

```python
# Def Function to Apply to GroupBy

def type_hs(df):
    
    # Calc for first 6 Columns in GroupBy Output
    
    district = df.iloc[1,8]
    tot_student = df['Student ID'].count()
    tot_bud = df.iloc[1,10]
    per_stud_bud=(tot_bud/tot_student)
    avg_math_sc=df['math_score'].mean()
    avg_read_sc=df['reading_score'].mean()
    
    #Percent Passing Math Calc
    
    index_fail = df[df['math_score']>=70].index
    math_pass = len(index_fail)
    per_math_pass ='{:,.2f}%'.format((math_pass/tot_student)*100)
    
     #Percent Passing Reading Calc
    
    index_fail = df[df['reading_score']>=70].index
    read_pass = len(index_fail)
    per_read_pass ='{:,.2f}%'.format((read_pass/tot_student)*100)
    
     #Percent Passing Math & Reading Calc
    
    index_pass = df[(df['reading_score']>=70) & (df['math_score']>=70)].index
    rw_pass = len(index_pass)
    per_rw_pass ='{:,.2f}%'.format((rw_pass/tot_student)*100)
    
    
    return pd.Series({'School Type':district,
                      'Total Students':'{:,.0f}'.format(tot_student),
                      'Total School Budget':'${:,.2f}'.format(tot_bud),
                      'Per Student Budget':'${:,.2f}'.format(per_stud_bud),
                      'Average Math Score':'{:,.2f}%'.format(avg_math_sc),
                      'Average Reading Score':'{:,.2f}%'.format(avg_read_sc),
                      '% Passing Math': per_math_pass,
                      '% Passing Reading': per_read_pass,
                      '% Overall Passing': per_rw_pass
                     })


# GroupBy Function

data_hs = data.groupby('school_name').apply(type_hs)
data_hs

```

```python
# ReGroup data_hs Without Format to Keep Values in Int and Float64


# Def Function to Apply to GroupBy

def type_hs(df):
    
    # Calc for first 6 Columns in GroupBy Output
    
    district = df.iloc[1,8]
    tot_student = df['Student ID'].count()
    tot_bud = df.iloc[1,10]
    per_stud_bud=(tot_bud/tot_student)
    avg_math_sc=df['math_score'].mean()
    avg_read_sc=df['reading_score'].mean()
    
    #Percent Passing Math Calc
    
    index_fail = df[df['math_score']>=70].index
    math_pass = len(index_fail)
    per_math_pass =(math_pass/tot_student)*100
    
     #Percent Passing Reading Calc
    
    index_fail = df[df['reading_score']>=70].index
    read_pass = len(index_fail)
    per_read_pass =(read_pass/tot_student)*100
    
     #Percent Passing Math & Reading Calc
    
    index_pass = df[(df['reading_score']>=70) & (df['math_score']>=70)].index
    rw_pass = len(index_pass)
    per_rw_pass =(rw_pass/tot_student)*100
    
    
    return pd.Series({'School Type':district,
                      'Total Students':tot_student,
                      'Total School Budget':tot_bud,
                      'Per Student Budget':per_stud_bud,
                      'Average Math Score':avg_math_sc,
                      'Average Reading Score':avg_read_sc,
                      '% Passing Math': per_math_pass,
                      '% Passing Reading': per_read_pass,
                      '% Overall Passing': per_rw_pass
                     })


# GroupBy Function

data_hs = data.groupby('school_name').apply(type_hs)
data_hs_df = pd.DataFrame(data_hs)

```

```python
# Assign data_hs Series to new DataFrame

data_hs_df = pd.DataFrame(data_hs)
data_hs_df

# Create Bins, Cut, and Assign Column

bins = [0,585,630,645,680]
group_names =  ['<$585','$585-630','$630-645','$645-680']

data_hs_df['Spending Range (Per Student)'] = pd.cut(data_hs_df['Per Student Budget'],bins,labels=group_names,include_lowest=True)
data_hs_df.head()

```

```python
# Call data_hs_df DataFrame and GroupBy Per Student Budget


spend_df = data_hs_df.groupby('Spending Range (Per Student)').agg({'Average Math Score':'mean',
                                                                   'Average Reading Score':'mean',
                                                                   '% Passing Math':'mean',
                                                                   '% Passing Reading':'mean',
                                                                   '% Overall Passing':'mean'
                                                                  })
pd.options.display.float_format = '{:.2f}%'.format
spend_df.head()
```
